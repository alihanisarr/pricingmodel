```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split


SF_listings_file = '/Users/alihanisar/Downloads/listingsSF.csv'
SFlistingsdata = pd.read_csv(SF_listings_file)
SFlistingsdata.describe()
SFlistingsdata.shape

SFlistingsdata = SFlistingsdata [["neighbourhood","latitude","longitude","room_type","price","minimum_nights","reviews_per_month"]]
SFlistingsdata.head()
SFlistingsdata = SFlistingsdata.dropna()
SFlistingsdata

y = SFlistingsdata.price

SFlistings_features = ["neighbourhood","room_type"]
X = pd.get_dummies(SFlistingsdata[SFlistings_features])

train_X, val_X, train_y, val_y = train_test_split( X , y , random_state=0 )

SFlistingsmodel = DecisionTreeRegressor(random_state=1)
SFlistingsmodel.fit(train_X , train_y)
SFmodelpredictions = SFlistingsmodel.predict(val_X)

mae = mean_absolute_error(val_y, SFmodelpredictions)
print("Decision Tree Mean Absolute Error:", mae)

SFlistingsdata.loc[val_X.index, "Predicted_price_DecisionTree"] = SFmodelpredictions



SFlistingsmodelRF = RandomForestRegressor(random_state=1)
SFlistingsmodelRF.fit(train_X , train_y)
SFmodelpredictionsRF = SFlistingsmodelRF.predict(val_X)

mae_RF = mean_absolute_error(val_y, SFmodelpredictionsRF)
print("Random Forest Mean Absolute Error:", mae_RF)

SFlistingsdata.loc[val_X.index, "Predicted_price_RandomForest"] = SFmodelpredictionsRF

SFlistingsdata















```
